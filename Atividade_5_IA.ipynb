###Realizando aproxima√ß√£o de fun√ß√µes com MLP 

####Arquivo de teste 1

```python
regr1 = MLPRegressor (hidden_layer_sizes = (4,4),
                     max_iter = 10000,
                     activation = 'identity',
                     solver = 'adam',
                     learning_rate = 'adaptive',
                     n_iter_no_change = 1000
)

m1 = []
for _ in range(50):
  regr1 = regr1.fit(x1,y1)
  m1.append(regr1.best_loss)
```

####Arquivo de teste 2

```python
regr2 = MLPRegressor (hidden_layer_sizes = (4,4),
                     max_iter = 10000,
                     activation = 'identity',
                     solver = 'adam',
                     learning_rate = 'adaptive',
                     n_iter_no_change = 1000
)

m2 = []
for _ in range(50):
  regr2 = regr2.fit(x2,y2)
  m2.append(regr2.best_loss_)

print(f'M√©dia de loss: {statistics.mean(m2)}')
print(f'Desvio padr√°o de loss: {statistics.pstdev(m2)}')
```

####Arquivo de teste 3

```python
regr3 = MLPRegressor (hidden_layer_sizes = (15,15),
                     max_iter = 10000,
                     activation = 'relu',
                     solver = 'adam',
                     learning_rate = 'adaptive',
                     n_iter_no_change = 1000
)

m3 = []
for _ in range(50):
  regr3 = regr3.fit(x3,y3)
  m3.append(regr3.best_loss_)

print(f'M√©dia de loss: {statistics.mean(m3)}')
print(f'Desvio padr√°o de loss: {statistics.pstdev(m3)}')
```

####Arquivo de teste 4

```python
regr4 = MLPRegressor (hidden_layer_sizes = (40,35),
                     max_iter = 10000,
                     activation = 'relu',
                     solver = 'adam',
                     learning_rate = 'adaptive',
                     n_iter_no_change = 1000
)

m4 = []
for _ in range(50):
  regr4 = regr4.fit(x4,y4)
  m4.append(regr4.best_loss_)

print(f'M√©dia de loss: {statistics.mean(m4)}')
print(f'Desvio padr√°o de loss: {statistics.pstdev(m4)}')
```

####Arquivo de teste 5

```python
regr5 = MLPRegressor (hidden_layer_sizes = (300, 300, 300, 300, 300),
                     max_iter = 10000,
                     activation = 'tanh',
                     solver = 'adam',
                     learning_rate = 'adaptive',
                     n_iter_no_change = 1000
)

m5 = []
for _ in range(5):
  regr5 = regr5.fit(x5,y5)
  m5.append(regr5.best_loss_)

print(f'M√©dia de loss: {statistics.mean(m5)}')
print(f'Desvio padr√°o de loss: {statistics.pstdev(m5)}')
```
